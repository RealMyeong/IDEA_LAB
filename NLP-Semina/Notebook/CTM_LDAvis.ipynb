```python
def text_normalize(text):
    key_list = word_dict.keys()
    for word in key_list:
        text = text.replace(word, word_dict[word])
    
    return text
```

```python
data['sample'] = data['sample'].apply(text_normalize)
data
```

```python
from contextualized_topic_models.models.ctm import CombinedTM
from contextualized_topic_models.utils.data_preparation import TopicModelDataPreparation
from contextualized_topic_models.utils.preprocessing import WhiteSpacePreprocessing
import nltk

import warnings

warnings.filterwarnings("ignore")

nltk.download('stopwords')
```

```python
documents = list(data['sample'].values)

sp = WhiteSpacePreprocessing(documents, stopwords_language='english', vocabulary_size=5000)
preprocessed_doc, unpre_doc, vocab, retained_indices = sp.preprocess()
```

```python
# ëª…ì‚¬ ì¶”ì¶œ
import re
import spacy
nlp = spacy.load('en_core_web_trf')

def extraction(text):
    text = re.sub('[^a-z]', ' ', text)
    text = re.sub('\\s+', ' ', text)
    text = text.strip()

    # Noun / Noun-Phrase Extraction ì§„í–‰
    doc = nlp(text)
    noun = [token.lemma_ for token in doc if token.pos_ == 'NOUN' and len(token.lemma_) >= 3]

    new_doc = ' '.join(noun)

    return new_doc

preprocessed_doc = [extraction(text) for text in preprocessed_doc]
```

```python
tp = TopicModelDataPreparation('paraphrase-distilroberta-base-v1')
training_dataset = tp.fit(text_for_contextual=unpre_doc, text_for_bow=preprocessed_doc)
```

```python
ctm = CombinedTM(bow_size=len(tp.vocab), contextual_size = 768, n_components=9, num_epochs=30)
ctm.fit(training_dataset)
```

```python
import pyLDAvis as vis

lda_vis_data = ctm.get_ldavis_data_format(tp.vocab, training_dataset, n_samples=40)

ctm_pd = vis.prepare(**lda_vis_data)
vis.display(ctm_pd)
```
