```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np

x = np.arange(1, 30, 0.1)

y1 = x
y2 = x * np.log(x)
y3 = x ** 2
y5 = 2 ** x

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y5)
plt.xlim([0,30])
plt.ylim([0,500])
plt.legend(['n', 'nlogn', 'n^2','2^n'])
plt.show()
```

```python
import pandas as pd

n = [10, 20, 50, 100, 1000]
n2 = [0.1 * (x ** 2) for x in n]
n3 = [0.1 * (x ** 2) + x + 100 for x in n]

df = pd.DataFrame(zip(n, n2, n3), columns=['n', '0.1n^2', '0.1n^2 + n + 100'])
df
```
